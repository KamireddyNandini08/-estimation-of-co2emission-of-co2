```python
data.drop(columns=['_id','Combined (mpg)','YEAR','VEHICLECLASS','FUELTYPE','Transmission','Model'],inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
encoded_cat_vars = encoder.fit_transform(data[['MAKE']])
```

```python
data.head()
```

```python
data.describe
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_num_vars = scaler.fit_transform(data[['ENGINESIZE','CYLINDERS','CITY','HIGHWAY','COMBINED']])
```

```python
scaled_num_vars
```

```python
X = np.concatenate((encoded_cat_vars.toarray(), scaled_num_vars), axis=1)
y = data['CO2EMISSIONS'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(64, activation='relu', input_shape=(59,)),  
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(8, activation='relu'),
    Dense(1)                                          
])

model.compile(optimizer='adam', loss='mean_squared_error')

history = model.fit(X_train, y_train, epochs=15, batch_size=32, validation_data=(X_test, y_test))

loss = model.evaluate(X_test, y_test)
print("Test Loss:", loss)
```

```python
predictions=model.predict(X_test)

# Print the predicted y values
print("Predicted y values:", predictions)
```

```python
y_test_array = np.array(y_test)
y_test_reshaped = y_test_array.reshape(-1, 1)
y_test_float32 = y_test_reshaped.astype(np.float32)
print(y_test_float32)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test_float32, predictions)
mae
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, predictions)
r2
```
